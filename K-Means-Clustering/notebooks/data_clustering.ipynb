```python
import sys
import os
sys.path.append(os.path.abspath('../src'))
from utils import load_dataset
from data_clustering import get_inertias_for_clustering, plot_inertia_k, get_cluster_model, visualize_clusters
```

```python
# Initializing variables
dataset_path = '../datasets/ready/mall_customers.csv'
k = 4
model_type = 'custom'
```

```python
#Loading the dataset
df = load_dataset(dataset_path)
```

```python
# finding the k useing elbow method
inertias = get_inertias_for_clustering(df)
print(inertias)
```

```python
# Plot the inertia vs k
plot_inertia_k(inertias)
```

```python
#clustering using k-means algorithm
model = get_cluster_model(df, k, model_type)
label = model.predict(df)
print(label)
df['label'] = label
```

```python
#Visualizing the clusters
visualize_clusters(df)
```
